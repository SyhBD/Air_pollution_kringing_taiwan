```python
air_pollution_index = air_pollution.copy()
```

```python
air_pollution_index = air_pollution_index.drop(columns=["ItemId","ItemName","ItemEngName","ItemUnit","MonitorMonth"])
```

```python
air_pollution_index = air_pollution_index.head(65)
```

```python
air_pollution_index
```

```python
for i in range(0,len(air_pollution_index)):
    target = air_pollution_index.iat[i,1]
    df = air_pollution[air_pollution.SiteName == target].Concentration.tolist()
    result = sum(df)/len(df)
    air_pollution_index.iat[i,2] = result
```

```python
air_pollution_index
```

```python
TWD97Lon=[]
TWD97Lat=[]
for i in range(0,len(air_pollution_index)):
    target = air_pollution_index.iat[i,1]
    flag = 0
    for j in range(0,len(station_code)):
        if station_code.iat[j,0] == target:
            TWD97Lon.append(station_code.iat[j,6])
            TWD97Lat.append(station_code.iat[j,7])
            flag = 1
        if flag == 0 and j == len(station_code)-1:
            print(target)
```

```python
air_pollution_index = air_pollution_index.drop(63)

air_pollution_index['TWD97Lon'] = TWD97Lon
air_pollution_index['TWD97Lat'] = TWD97Lat
```

```python
air_pollution_index
```

```python
import pykrige.kriging_tools as kt
from pykrige.ok import OrdinaryKriging
```

```python
ok = OrdinaryKriging(
    air_pollution_index.TWD97Lon.tolist(),
    air_pollution_index.TWD97Lat.tolist(),
    air_pollution_index.Concentration.tolist(),
    variogram_model="linear",
    verbose=False,
    enable_plotting=False,
)
```

```python
ok
```

```python
ok.execute("grid",121.519884,25.032405)[0][0][0]
```

```python
predict_list = []
for i in range(0,len(district_code)):
    x = district_code.iat[i,2]
    y = district_code.iat[i,3]
    predict_list.append(ok.execute("grid",x,y)[0][0][0])
```
