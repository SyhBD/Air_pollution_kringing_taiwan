```python
import pandas as pd
import numpy as np
```

```python
station_code = pd.read_excel("../空氣品質監測站基本資料(含地理圖資).xlsx")
```

```python
air_pollution = pd.read_excel("../空氣品質監測月值_mod3.xlsx")
```

```python
air_pollution
```

```python
air_pollution = air_pollution[air_pollution.Concentration != "x"]
```

```python
district_code = pd.read_excel("../行政區中心點_mod.xlsx")
```

```python
district_code
```

```python
TWD97Lon=[]
TWD97Lat=[]
for i in range(0,len(air_pollution)):
    target = air_pollution.iat[i,1]
    flag = 0
    for j in range(0,len(station_code)):
        if station_code.iat[j,0] == target:
            TWD97Lon.append(station_code.iat[j,6])
            TWD97Lat.append(station_code.iat[j,7])
            flag = 1
        if flag == 0 and j == len(station_code)-1:
            print(target)
```

```python
air_pollution = air_pollution[air_pollution.SiteName != "富貴角"]
```

```python
air_pollution
```

```python
air_pollution['TWD97Lon'] = TWD97Lon
air_pollution['TWD97Lat'] = TWD97Lat
```

```python
air_pollution
```

```python
ap_group = air_pollution.groupby("MonitorMonth")
```

```python
predict_result = district_code.copy()
predict_result = predict_result.drop(columns=["行政區名","TGOS_URL"])
```

```python
predict_result = pd.concat([predict_result]*120).reset_index()
```

```python
year = []
month = []
time = []
for i in ap_group.groups.keys():
    for j in range(0,371):
        year.append(int(i / 100))
        month.append(i % 100)
        time.append(i)
```

```python
predict_result["年"] = year
predict_result["月"] = month
predict_result["時間"] = time
```

```python
predict_result
```

```python
import pykrige.kriging_tools as kt
from pykrige.ok import OrdinaryKriging
import matplotlib.pyplot as plt
```

```python
predict_list = []
for i in ap_group.groups.keys():
    
    ap = ap_group.get_group(i)
    
    ok = OrdinaryKriging(
    ap.TWD97Lon.tolist(),
    ap.TWD97Lat.tolist(),
    ap.Concentration.tolist(),
    variogram_model="linear",
    verbose=False,
    enable_plotting=False,
    )
    
    for i in range(0,len(district_code)):
        x = district_code.iat[i,2]
        y = district_code.iat[i,3]
        predict_list.append(ok.execute("grid",x,y)[0][0][0])
        
```

```python
len(predict_list)
```

```python
predict_result['predict_PM2.5'] = predict_list
```

```python
predict_result.to_excel("../PM2.5行政區預測20200909_1.xlsx")
```
